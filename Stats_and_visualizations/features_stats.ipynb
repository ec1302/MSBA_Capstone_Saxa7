```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ignoring warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('na_filled_df.csv')
df = df.sample(n=35000, random_state=2025)
```

```python
# Create a bar chart for 'derived_race' counts
plt.figure(figsize=(10, 6))
sns.countplot(x='derived_race', data=df)
plt.title('Counts of each category in derived_race')
plt.xticks(rotation=45)
plt.show()
```

```python
# Table of above data
print("Counts of each category in 'derived_race':")
print(df['derived_race'].value_counts())
```

```python
# Create a bar chart for 'applicant_sex' counts
plt.figure(figsize=(10, 6))
sns.countplot(x='applicant_sex', data=df)
plt.title('Counts of each category in applicant_sex')
plt.xticks(rotation=45)  
plt.show()
```

```python
# Table of above data
print("Counts of each category in 'applicant_sex':")
print(df['applicant_sex'].value_counts())
```
