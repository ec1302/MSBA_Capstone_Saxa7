```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.api.types import CategoricalDtype
```

```python
# Smaller data frame of 50,000 rows
df = pd.read_csv('sampled_data_50k.csv')
```

```python
columns_to_select = ['action_taken', 'conforming_loan_limit', 'derived_loan_product_type', 'derived_dwelling_category', 'preapproval', 
                     'loan_purpose', 'lien_status', 'reverse_mortgage', 'open-end_line_of_credit', 'business_or_commercial_purpose', 'loan_amount', 'loan_to_value_ratio',
                      'rate_spread', 'total_loan_costs', 'loan_term', 'negative_amortization', 'occupancy_type', 'income', 'debt_to_income_ratio']
df = df[columns_to_select]
df.shape
```

```python
# Setting the style
sns.set_style("whitegrid")

# For each variable, plot histograms or bar plots and display statistics
for col in columns_to_select:
    if col != 'action_taken' and col in df.columns:
        # Check the datatype for the column
        column_dtype = df[col].dtypes
        
        if column_dtype in ['int64', 'float64']:  # if numeric
            plt.figure(figsize=(10, 6))
            sns.histplot(df[col], bins=30, kde=True)
            plt.title(f'Histogram for {col}')
            plt.show()

            # Print statistics
            print(f'Mean of {col}: {df[col].mean()}')
            print(f'Median of {col}: {df[col].median()}')
            print(f'Mode of {col}: {df[col].mode().iloc[0]}')
            print('------' * 10)
        
        elif isinstance(column_dtype, CategoricalDtype):  # if categorical
            plt.figure(figsize=(10, 6))
            sns.countplot(data=df, y=col)
            plt.title(f'Bar plot for {col}')
            plt.show()

            # Print mode
            print(f'Mode of {col}: {df[col].mode().iloc[0]}')
            print('------' * 10)
```
